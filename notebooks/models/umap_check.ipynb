```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
import pandas as pd
import umap
```

```python
cond_dir = '../../data/intermediates/pca/cond1_pat1to81_outrmv_standard'
X_train = np.loadtxt(cond_dir + '/X_train_cond1_pat1to81_outrmv_standard.csv', delimiter=',')
X_test = np.loadtxt(cond_dir + '/X_test_cond1_pat1to81_outrmv_standard.csv', delimiter=',')
y_train = pd.read_csv(cond_dir + '/y_train_cond1_pat1to81_outrmv_standard.csv')
y_test = pd.read_csv(cond_dir + '/y_test_cond1_pat1to81_outrmv_standard.csv')
```

```python
X_pca = np.concatenate([X_train, X_test], axis=0)
y = np.concatenate([y_train['class'].values, y_test['class'].values])
print(X_pca.shape)
print(y.shape)
```

```python
# use UMAP to map the 60 principal components into 2
reducer = umap.UMAP(n_components=2, random_state=42)
embeddings = reducer.fit_transform(X_pca)
```

```python
cmap = ListedColormap(['blue', 'orange'])

plt.scatter(embeddings[:, 0],embeddings[:, 1], c=y, cmap=cmap)
plt.xlabel('UMAP-1')
plt.ylabel('UMAP-2')
```
