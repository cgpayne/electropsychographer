```python
import pandas as pd
import matplotlib.pyplot as plt
import csv

plt.rc('text', usetex=False)
# plt.rc('text', usetex=True)
```

```python
data_dir = '../../data/pruned'
meta_dir = '../../data/meta_archive'

with open(meta_dir + '/columnLabels.csv', 'r') as fin:
    heading = list(csv.reader(fin))[0]

print(heading, '\n')

dfHC = pd.read_csv(data_dir + '/1.csv', header=None)
dfHC.columns = heading
display(dfHC)

dfSZ = pd.read_csv(data_dir + '/81.csv', header=None)
dfSZ.columns = heading
display(dfSZ)
```

```python
dfHC_1 = dfHC[dfHC['condition'] == 1]
dfHC_2 = dfHC[dfHC['condition'] == 2]
dfHC_3 = dfHC[dfHC['condition'] == 3]

dfSZ_1 = dfSZ[dfSZ['condition'] == 1]
dfSZ_2 = dfSZ[dfSZ['condition'] == 2]
dfSZ_3 = dfSZ[dfSZ['condition'] == 3]
```

```python
electrodes = heading.copy()
electrodes.remove('subject')
electrodes.remove('trial')
electrodes.remove('condition')
electrodes.remove('sample')
print(electrodes)
```

```python
def normalize_EEG(df, electrodes):
    df_tmp = (df - df.min())/(df.max() - df.min())
    df[electrodes] = df_tmp[electrodes]
```

```python
normalize_EEG(dfHC_1, electrodes)
normalize_EEG(dfHC_2, electrodes)
normalize_EEG(dfHC_3, electrodes)
normalize_EEG(dfSZ_1, electrodes)
normalize_EEG(dfSZ_2, electrodes)
normalize_EEG(dfSZ_3, electrodes)
```

```python
def plot_EEG(dfHC_1, dfHC_2, dfHC_3, dfSZ_1, dfSZ_2, dfSZ_3, electrode):
    plt.figure()
    plt.plot(dfHC_1['sample'], dfHC_1[electrode], label='HC_1')
    plt.plot(dfHC_2['sample'], dfHC_2[electrode], label='HC_2')
    plt.plot(dfHC_3['sample'], dfHC_3[electrode], label='HC_3')
    plt.plot(dfSZ_1['sample'], dfSZ_1[electrode], label='SZ_1')
    plt.plot(dfSZ_2['sample'], dfSZ_2[electrode], label='SZ_2')
    plt.plot(dfSZ_3['sample'], dfSZ_3[electrode], label='SZ_3')
    plt.legend()
    plt.title(electrode)
    plt.show()
    plt.close()

def plot_per_cond(dfHC, dfSZ, cond, electrode):
    plt.figure()
    plt.plot(dfHC['sample'], dfHC[electrode], label='HC')
    plt.plot(dfSZ['sample'], dfSZ[electrode], label='SZ')
    plt.legend()
    plt.title(f"HC vs SZ, condition: {cond}, electrode: {electrode}")
    plt.xlabel("time [ms]")
    plt.ylabel("normalized amplitude")
#     plt.title(f"HC vs SZ, cond: {cond} , electrode: {electrode}", fontsize=16)
#     plt.xlabel(r"time [$ms$]", fontsize=14)
#     plt.ylabel(f"amplitude [$\mu V$]", fontsize=14)
    plt.show()
    plt.close()

def plot_diff(dfHC_A, dfHC_B, dfSZ_A, dfSZ_B, flag, electrode):
    plt.figure()
    plt.plot(dfHC_A['sample'], dfHC_A[electrode].values - dfHC_B[electrode].values, label='HC')
    plt.plot(dfSZ_A['sample'], dfSZ_A[electrode].values - dfSZ_B[electrode].values, label='SZ')
    plt.legend()
    plt.title(f"difference of condition {flag}, electrode: {electrode}")
    plt.xlabel("time [ms]")
    plt.ylabel("normalized amplitude")
#     plt.title(f"diff of {flag}, electrode: {electrode}", fontsize=16)
#     plt.xlabel(r"time [$ms$]", fontsize=14)
#     plt.ylabel(f"amplitude [$\mu V$]", fontsize=14)
    plt.show()
    plt.close()

def plot_vs_cond(df_1, df_2, df_3, pat, electrode):
    plt.figure()
    plt.plot(df_1['sample'], df_1[electrode], label='cond 1')
    plt.plot(df_2['sample'], df_2[electrode], label='cond 2')
    plt.plot(df_3['sample'], df_3[electrode], label='cond 3')
    plt.legend()
    plt.title(f"condition 1 vs 2 vs 3, patient: {pat}, electrode: {electrode}")
    plt.xlabel("time [ms]")
    plt.ylabel("normalized amplitude")
#     plt.title(f"1 vs 2 vs 3, pat: {pat}, electrode: {electrode}", fontsize=16)
#     plt.xlabel(r"time [$ms$]", fontsize=14)
#     plt.ylabel(f"amplitude [$\mu V$]", fontsize=14)
    plt.show()
    plt.close()
```

```python
# left-front, right-front, left-mid, right-mid, left-back, right-back
auditory_cortex = ['T7', 'T8', 'TP7', 'TP8', 'P7', 'P8']
```

```python
%%time
# for electrode in ['Fp1', 'AF7', 'AF3', 'F1', 'F3', 'F5', 'F7', 'FT7', 'FC5']:
# for electrode in electrodes:
for electrode in auditory_cortex:
    plot_per_cond(dfHC_1, dfSZ_1, '1', electrode)
    plot_per_cond(dfHC_2, dfSZ_2, '2', electrode)
    plot_per_cond(dfHC_3, dfSZ_3, '3', electrode)
```

```python
%%time
# for electrode in ['Fp1', 'AF7', 'AF3', 'F1', 'F3', 'F5', 'F7', 'FT7', 'FC5']:
# for electrode in electrodes:
for electrode in auditory_cortex:
    plot_diff(dfHC_1, dfHC_2, dfSZ_1, dfSZ_2, "1 - 2", electrode)
    plot_diff(dfHC_1, dfHC_3, dfSZ_1, dfSZ_3, "1 - 3", electrode)
    plot_diff(dfHC_2, dfHC_3, dfSZ_2, dfSZ_3, "2 - 3", electrode)
```

```python
%%time
# for electrode in ['Fp1', 'AF7', 'AF3', 'F1', 'F3', 'F5', 'F7', 'FT7', 'FC5']:
# for electrode in electrodes:
for electrode in auditory_cortex:
    plot_vs_cond(dfHC_1, dfHC_2, dfHC_3, 'HC', electrode)
    plot_vs_cond(dfSZ_1, dfSZ_2, dfSZ_3, 'SZ', electrode)
```
